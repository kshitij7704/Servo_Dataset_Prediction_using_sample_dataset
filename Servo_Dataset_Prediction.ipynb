```python
import pandas as pd
import numpy as np
```

```python
servo_data = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/Servo%20Mechanism.csv", on_bad_lines = "skip")
```

```python
servo_data.head()
```

```python
servo_data.describe()
```

```python
servo_data.columns
```

```python
servo_data.shape
```

```python
servo_data[['Motor']].value_counts()
```

```python
servo_data[['Screw']].value_counts()
```

```python
servo_data.replace({'Motor': {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}}, inplace = True)
```

```python
servo_data.replace({'Screw': {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4}}, inplace = True)
```

```python
y = servo_data['Class']
```

```python
y.shape
```

```python
x = servo_data[['Motor', 'Screw', 'Pgain', 'Vgain']]
```

```python
x.shape
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
x_train.size, x_test.size, y_train.size, y_test.size
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```
